### 단축 평가 (short-circuit evaluation)

 - 단축 평가는 계산을 진행하는 도중 결과가 이미 확정된 경우, 나머지 계산을 생략하는 것을 말함
 
 - and 연산과 or 연산은 특성상 단축 평가를 할 수 있다
 
 - 컴퓨터의 계산을 하기 위한 자원(리소스; resource)을 절약하기 위하여 사용
 
 - 하지만 if문을 사용하는 것이 코드는 길어지지만 더 읽고 이해하기가 쉬움을 알아야 함

```python
# and 연산의 단축 평가

(1 == 2) and (3 < 4) and (True != False)
```

```python
# or 연산의 단축 평가

('red' in ['red', 'violet', 'blue']) or (1 == 10 - 9) and (20 < 20)
```

```python
# and 와 or 를 이용한 선택

def 홀짝(n):
    """n이 홀수인지 짝수인지를 화면에 출력한다."""
    (n % 2 == 0) and print(n, '은 짝수입니다.')
    (n % 2 == 0) or print(n, '은 홀수입니다.')

홀짝(10)
홀짝(11)
```

### 재귀 (recursion)

 - 함수가 자신을 직접 또는 간접적으로 호출하는 것을 **재귀(recursion)**
 
 - 재귀를 사용하면 while 문이나 for 문 없이도 반복 작업을 수행할 수 있음

```python
# n 이상 m 미만의 자연수 출력

def 자연수(n, m):
    """수 n을 출력하고, 1 더한 수가 m보다 작으면 그 수도 출력한다."""
    print(n)
    if n + 1 < m:
        자연수(n + 1, m)

자연수(4, 8)
```

 - 자연수() 함수는 자연수 n과 m을 입력받아, 그 사이의 모든 자연수를 출력하는 함수
 
 - 이 함수는 본문 안에서 자기자신(자연수())을 호출
 
 - 이 함수가 실행되는 과정을 생각해 보자

> 1. 처음에 자연수(4, 8)을 호출. 화면에 4가 출력된다. 4 + 1 < 8는 참이므로, 자연수(4 + 1, 8)이 재귀 호출된다.
>
> 2. 다시 호출된 자연수()함수는 전달받은 수 n을 출력하고, 조건을 검사해 자연수(n + 1, m)을 다시 호출한다.
>
> 3. 자연수() 함수는 계속 동일한 형태로 호출되지만, 이 함수의 매개변수 n에 전달되는 값은 계속 변한다. n + 1 < m 이라는 조건이 거짓이 된 호출에서 재귀 호출은 더이상 일어나지 않는다.

```python
# 재귀를 이용한 피보나치 수 계산

def n번째_피보나치_수(n):
    """n번째 피보나치 수를 반환한다."""
    if n == 1:     # 첫 번째 피보나치 수는 1
        return 1
    elif n == 2:   # 두 번째 피보나치 수는 1
        return 1
    else:          # 그 후의 피보나치 수
        return n번째_피보나치_수(n - 1) + n번째_피보나치_수(n - 2)

# 1번째부터 11번째 피보나치 수까지 출력
for n in range(1, 12):
    print(n번째_피보나치_수(n), end=' ')
```

 - 재귀는 특정한 문제를 해결할 때 편리하지만 단점도 있다. 
 
 - 재귀는 함수의 실행이 끝나기도 전에 또 다시 함수를 실행하기 때문에, 재귀 호출이 연달아 많이 일어나면 컴퓨터의 계산 자원과 기억 자원을 빠르게 소모할 수 있음
 
 - 위에서 정의한 자연수() 함수를 이용해 큰수를 출력하도록 해 보면 오류가 발생할 수 있음
