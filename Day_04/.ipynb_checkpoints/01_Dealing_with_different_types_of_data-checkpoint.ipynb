### bit & byte

 - 컴퓨터는 전기로 동작한다
 
 - 고로 컴퓨터는 전기를 사용하여 데이터를 저장하고 다룰 것이다 --> 어떻게?

 - 전기의 값이 높으면 1, 낮으면 0으로 기억하게 제작되어진 컴퓨터 내의 부품들
 
 - 0과 1의 여러 조합으로 데이터들을 표현 --> 즉, 정보를 가지고 있다
 
 - 이렇게 0과 1로 표현되는 정보의 단위를 bit 비트라고 한다
 
 - 0과 1의 조합을 묶음으로 (조합으로) 기억한다면 더 많은 정보를 표기할 수 있다
 
 - 8개의 비트를 사용한 묶음 정보 단위를 byte 바이트라고 한다

 - **bit : 두가지 상태를 구별할 수 있는 정보의 단위**
 - **byte : bit 8개를 묶은 정보의 단위. 총 256개의 상태를 구별할 수 있음**

### 실수

 - 소수점이 표현되는 수이다
 
 - python에서는 float라는 데이터 유형을 사용한다

```python
3.141592
```

 - 과학 표기법 : e = 10과 같이 취급하여 큰 수 혹은 여러 자리를 필요로하는 수를 간단하게 표현 할 수 있음

```python
1.23e0   # 1.23 * 10 ^ 0
```

```python
1.23e6   # 1.23 * 10 ^ 6
```

```python
1.23e-4  # 1.23 * 10 ^ -4
```

 - 실수는 정밀도의 문제를 가진다
 
 - 정확하게 계산되어야할 것이 오차를 가질 수 있다

```python
1.1 - 1.0         # 계산 결과는 0.1이어야 하지만...
```

```python
# 1.1 - 1.0 == 0.1 --> X
if(1.1 - 1.0 == 0.1):
    print(True)
```

```python
# 0.00009 < 1.1 - 1.0 < 0.100001 --> O
if(0.00009 < 1.1 - 1.0 < 0.100001):
    print(True)
```

 - 무한대를 표현할 수 있음

```python
float('inf')          # 양의 무한대
```

```python
float('-inf')         # 음의 무한대
```

```python
1e100 < float('inf')  # 무한대는 다른 모든 실수보다 크다
```

```python
1e309                 # 너무 큰 실수는 무한대로 평가된다
```

```python
1e-324                # 너무 0에 가까운 실수는 0으로 평가된다
```

### 복소수 complex number

 - 실수와 허수를 가지는 수

```python
1-2j         # 실수부가 1, 허수부가 -2인 복소수
```

```python
(1-2j).real  # 복소수의 실수부
```

```python
(1-2j).imag  # 복소수의 허수부
```

### 수의 여러 타입과 함께하는 연산

 - 연산은 계산!
 
 - 연산자로 +, - 등등 말고도 '수'라는 항목에서 공유하는 것들이 많음 --> 정수든 실수든 반올림은 필요하고 곱하기도 할 수 있으니까

```python
round(10)  # 실수와 마찬가지로 정수도 반올림이 된다
```

```python
(10+3j).imag  # 정수에서 복소수의 실수부를 구하는 연산도 가능하다
```

```python
10 + 0.5    # 정수와 실수의 연산 -> 실수
```

```python
0.5 * 1+2j  # 실수와 복소수의 연산 -> 복소수
```

```python
(10).upper()  # 텍스트를 위한 연산은 오류를 발생시킨다
```
